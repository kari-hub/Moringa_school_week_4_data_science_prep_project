# 1.5 Cleaning our datasets

We will then proceed to perform the data cleaning procedures needed for the dataset given.



```
# Removing irrelevant Data

data_df.drop(['Displayed comment','Scheduled at', 'Slots', 'Station type', 'Status', 'Subscription status'], axis=1, inplace=True)
data_df
```

```
# Check for any missing values

data_df.isnull().values.any()

# Get the sum of the null values

data_df.isnull().sum().sum()

# Drop all rows with null values.
df1 = data_df.dropna(how='all')
df1

# Check for duplicates
data_df.duplicated()

# Change column names
df = df1.rename(columns= {'Bluecar counter' : 'Bluecar',	'Utilib counter' : 'Utilib', 'Utilib 1.4 counter' : 'Utilib 1.4'})
df

# Fill null values with 0
df2 = df.fillna(0)
df2

# We set the limit for the number of non-null values.
df2_thresh = df2.dropna(thresh=6)
df2
```

```
# We will remove all existing syntax errors

df['City']=df['City'].str.strip()
df['ID']= df['ID'].str.strip()
df['Kind']=df['Kind'].str.strip()
df['Public name']=df['Public name'].str.strip()
df['Rental status']=df['Rental status'].str.strip()
df['Status']=df['Status'].str.strip()
df


# We will replace any other errors in our data using the str function

df.columns = df.columns.str.strip().str.upper().str.replace("","")


# We will then filter out the information that we will need for the survey
# By filtering the rows first.

df = df2[df2['RENTAL STATUS']=="operational"]
df

df = df2[df2['STATUS']=="ok"]
df

df = df2[df2['KIND']=="STATION"]
df

# We will then remove rows with 0 value for bluecar, utilib and utilib 1.4

df = df[df['UTILIB 1.4'] > 0]
df = df[df['UTILIB'] > 0]
df = df[df['BLUECAR'] > 0]
df
```
